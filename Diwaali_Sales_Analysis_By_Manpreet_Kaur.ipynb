## Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = data1)
```

```python
#This code groups your DataFrame 'data' by the 'Gender' column and 
# calculates the sum of the 'Amount' column for each group, resulting in the total sales for each gender.


sales_generated = data1.groupby(['Gender'])['Amount'].agg('sum')
                         
sales_generated 
```

```python
# This code first sorts the 'sales_generated' DataFrame by the 'Amount' column in descending order.
# Then, it resets the index to make it a proper DataFrame with the 'Gender' and 'Amount' columns. 
# This way, 'sales_gen' will contain the sorted data with 'Gender' and 'Amount'.




sales_gen=sales_generated.sort_values(by='Amount', ascending=False)

sales_gen = sales_generated.reset_index().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
# plotting a bar chart for gender vs total amount

sns.barplot(x='Gender', y='Amount', data=sales_gen)

plt.show()
```

```python

# Count the number of occurrences of each gender
gender_counts = data1['Gender'].value_counts()

# Create a pie chart
plt.pie(gender_counts, labels=gender_counts.index, colors=['lightcoral', 'lightblue'])

# Set the title (optional)
plt.title("Gender Distribution")

plt.show()
```

### From the above analysis, we can say that women have more purchasing power and doing more shopping as compared to men
